```python
def Birthday(n):
    import numpy as np
    days = []
    r = np.random.rand(n) 
    print r
    for i in range(365):
        days.append(False) 
    bd = False
    for i in range(n):
        number = int(364*r[i] + 1) 
        print number
        if (days[number]):
            bd = True
            break
        days[number] = True
    return bd
```

```python
Birthday(200)
```

```python
def Probably(n, npts):
    sum = 0.0
    for i in range(npts):
        if(Birthday(n)):
            sum = sum + 1.0
    return sum/npts
```

```python
import numpy as np
for i in np.arange(5,200,5):
    pr = Probably(i,3000)
    print(repr(pr))
```
